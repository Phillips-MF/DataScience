```python
import pandas as pd
import apyori
```

```python
grocery_products = pd.read_csv("Grocery Products Purchase.csv", header = 0)
```

# Dataframe analysis

```python
len(grocery_products)
```

```python
for index in grocery_products.columns:
    grocery_products[index] = grocery_products[index].str.strip()#Take out backspace
```

```python
products = grocery_products.values.flatten()
products_list = [product for product in products if pd.notna(product)]
products_unique = list(set(products_list))
products_unique_sorted = sorted(products_unique)
```

```python
print(f'There are {len(products_unique_sorted)} itens\n  {products_unique_sorted}')
```

# Apriori

```python
grocery_qtd = grocery_products.notna().apply(sum, axis=1)
```

```python
grocery_qtd
```

```python
bag = [set(line.dropna()) for _, line in grocery_products[grocery_qtd > 1].iterrows()]
bag[:10]
```

```python
len(bag)
```

```python
minsup = 0.04
minconf = 0.3

association_rules = apyori.apriori(bag, min_support=minsup, min_confidence = minconf)

for rule in association_rules:
    items = list(rule.items)
    print(f'{items[:-1]} --> [{items[-1]}] Support: {rule.support:.3f}'
          f' Confidence: {rule.ordered_statistics[0].confidence:.3f}'      
          )
```
